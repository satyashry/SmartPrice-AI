```python
# cleaning prices and converting them into float/int
def clean_price(prices):
    cleaned_list = []
    for price in prices:
        if pd.isnull(price):
            cleaned_list.append(None)
        else:
            cleaned_list.append(float(str(price).replace("₹","").replace(",","")))
    return cleaned_list 

data["current_price"] = clean_price(data["current_price"])
data["original_price"] = clean_price(data["original_price"])
```

## Extracting Ram and other specification

```python
import re # regex
```

```python
def extract_ram(text):
    match = re.search(r'(\d+)\s*GB', str(text))
    return int(match.group(1)) if match else None

# apply row by row
data["ram"] = data["brand"].apply(extract_ram)
print(data["ram"].value_counts())
```

```python
data["ram"] = data["brand"].apply(extract_ram)
```

## More columns to extract 


col : brand , processor , storage, graphics, os, product name

```python
def extract_storage(text):
    # handle TB first
    match_tb = re.search(r'(\d+)\s*TB', str(text))
    if match_tb:
        return int(match_tb.group(1)) * 1024  # convert to GB
    
    # then GB SSD
    match_gb = re.search(r'(\d+)\s*GB\s*SSD', str(text))
    if match_gb:
        return int(match_gb.group(1))
    
    return None

data["storage"] = data["brand"].apply(extract_storage)
print(data["storage"].value_counts())
```

```python
df.isnull().sum()
```

```python
def extract_os(text):
    match = re.search(r'(Windows \d+|DOS|Ubuntu)', str(text))
    return match.group(1) if match else None

data["os"] = data["brand"].apply(extract_os)
```

```python
def extract_graphics(text):
    match = re.search(r'(\d+\s*GB\s*Graphics|NVIDIA|AMD Radeon|Intel Iris)', str(text))
    return match.group(1) if match else "Integrated"
```

```python
def extract_processor(text):
    match = re.search(r'(Intel Core [^\s]+|AMD Ryzen \d+|Snapdragon [^\s]+|Apple M\d)', str(text))
    return match.group(1) if match else None
```

```python
def extract_brand(text):
    return str(text).split()[0]
```

```python
data["graphics"]     = data["brand"].apply(extract_graphics)
data["processor"]    = data["brand"].apply(extract_processor)
data["brand"]   = data["brand"].apply(extract_brand)
```

```python
def fill_value(df):
    for col in df.columns :
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = df[col].fillna(df[col].median())
        else :
            df[col] = df[col].fillna(df[col].mode()[0])
        
    return df
    
```

```python
df = fill_value(df)
```
